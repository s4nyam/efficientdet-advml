```python
import csv
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def swish(x):
    return x / (1 + np.exp(-x))

def d_relu(x):
    return np.where(x > 0, 1, 0)

def d_swish(x):
    return swish(x) + (1 - swish(x)) * (x * np.exp(-x)) / (1 + np.exp(-x))

# Define x values
x = np.arange(-5, 5, 0.1)

# Calculate y values for each function
relu_y = relu(x)
swish_y = swish(x)
d_relu_y = d_relu(x)
d_swish_y = d_swish(x)

# Save data to CSV files
with open('relu.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x', 'ReLU'])
    for i in range(len(x)):
        writer.writerow([x[i], relu_y[i]])

with open('swish.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x', 'Swish'])
    for i in range(len(x)):
        writer.writerow([x[i], swish_y[i]])

with open('d_relu.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x', 'dReLU'])
    for i in range(len(x)):
        writer.writerow([x[i], d_relu_y[i]])

with open('d_swish.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x', 'dSwish'])
    for i in range(len(x)):
        writer.writerow([x[i], d_swish_y[i]])

# Plot functions
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0, 0].plot(x, relu_y, label='ReLU')
axs[0, 0].plot(x, swish_y, label='Swish')
axs[0, 0].set_title('ReLU and Swish')
axs[0, 0].legend()
axs[0, 1].plot(x, d_relu_y, label='dReLU')
axs[0, 1].plot(x, d_swish_y, label='dSwish')
axs[0, 1].set_title('Derivative ReLU and Derivative Swish')
axs[0, 1].legend()
axs[1, 0].plot(x, relu_y, label='ReLU')
axs[1, 0].plot(x, swish_y, label='Swish')
axs[1, 0].set_xlim([-2, 2])
axs[1, 0].set_ylim([-1, 2])
axs[1, 0].set_title('ReLU and Swish (Zoomed in)')
axs[1, 0].legend()
axs[1, 1].plot(x, d_relu_y, label='dReLU')
axs[1, 1].plot(x, d_swish_y, label='dSwish')
axs[1, 1].set_xlim([-2, 2])
axs[1, 1].set_ylim([-0.5, 1.5])
axs[1, 1].set_title('Derivative ReLU and Derivative Swish (Zoomed in)')
axs[1, 1].legend()
fig.suptitle('Activation Functions and their Derivatives')
fig.tight_layout()
fig.savefig('activations.png')
plt.show()
```
